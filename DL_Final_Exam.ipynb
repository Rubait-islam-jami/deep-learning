## Design and implement a deep Convolutional Neural Network (CNN) for a multi-class image classification problem using PyTorch. The network should consist of four convolutional blocks, where each block includes two convolution layers followed by activation and normalization, and a pooling layer at the end. The number of filters should increase progressively across blocks. After the convolutional part, flatten the feature maps and connect them to fully connected layers with dropout for regularization, and produce final outputs corresponding to K classes. Also explain the role of each part of the network in detail.

# Write Answer 10:

```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import transforms
from torch.utils.data import Dataset,DataLoader,Subset
from PIL import Image
import kagglehub

device= torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print("Device:",device)

path = kagglehub.dataset_download("mohitsingh1804/plantvillage")

Train_path= os.path.join(path,"PlantVillage","train")
VAL_path= os.path.join(path,"PlantVillage","val")

transform = transforms.Compose([
    transforms.Resize((128,128)),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3,[0.5]*3)
])

class PlantDataset(Dataset):
      def __init__(self,root_dir,transform=None):
         self.samples=[]
         self.transform=transform

         self.classes = sorted([
              d for d in os.listdir(root_dir)
              if os.path.isdir(os.path.join(root_dir,d))
         ])
         self.class_to_idx={
             cls_name: i for i,cls_name in enumerate(self.classes)
         }

         for cls in self.classes:
             class_path=os.path.join(root_dir,cls)

             for img_name in os.listdir(class_path):
                 img_path=os.path.join(class_path,img_name)

                 self.samples.append((
                     img_path,self.class_to_idx[cls]))
      def __len__(self):
          return len(self.samples)
      def __getitem__(self,idx):
          img_path,label = self.samples[idx]

          image = Image.open(img_path).convert("RGB")

          if self.transform:
             image= self.transform(image)
             return image,label

train_dataset = PlantDataset(Train_path,transform=transform)
val_dataset = PlantDataset(VAL_path,transform=transform)


train_loader = DataLoader(
    Subset(train_dataset,range(500)),
    batch_size=32,
    shuffle=True
)
val_loader = DataLoader(
    Subset(val_dataset,range(100)),
    batch_size=32,
    shuffle=False
)

class DeepPlantCNN(nn.Module):
    def __init__(self, num_classes):
         super(DeepPlantCNN, self).__init__()

         self.block1 = self._make_block(3, 32)
         self.block2 = self._make_block(32, 64)
         self.block3 = self._make_block(64, 128)
         self.block4 = self._make_block(128, 256)

         self.classifier = nn.Sequential(
             nn.Flatten(),
             nn.Linear(256*8*8,512),
             nn.ReLU(),
             nn.Dropout(0.5),
             nn.Linear(512,num_classes)
        )

    def _make_block(self,in_c,out_c):
        return nn.Sequential(
        nn.Conv2d(in_c,out_c,3,padding=1),
        nn.BatchNorm2d(out_c),
        nn.ReLU(),

        nn.Conv2d(out_c,out_c,3,padding=1),
        nn.BatchNorm2d(out_c),
        nn.ReLU(),

        nn.MaxPool2d(2)

    )
    def forward(self,x):
      x= self.block1(x)
      x= self.block2(x)
      x= self.block3(x)
      x= self.block4(x)
      x= self.classifier(x)
      return x



num_classes = len(train_dataset.classes)
model =DeepPlantCNN(num_classes).to(device)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(),lr=0.001)


epochs=5

for epoch in range(epochs):
   model.train()
   total_loss=0

   for images, labels in train_loader:
       images,labels = images.to(device),labels.to(device)

       optimizer.zero_grad()
       outputs = model(images)

       loss= criterion(outputs,labels)
       loss.backward()
       optimizer.step()

       total_loss+=loss.item()

   print(f"Epoch[{epoch+1}/epochs] Loss: {total_loss/len(train_loader):.4f}")

model.eval()
correct=0
total =0

with torch.no_grad():
  for images,labels in val_loader:
    images,labels= images.to(device),labels.to(device)
    outputs=model(images)
    _,predicted = torch.max(outputs,1)

    total+=labels.size(0)
    correct+=(predicted==labels).sum().item()

print("Accuracy:",100* correct/total)

```

Role of each part:       

1.convolution Layer: the convulation layer extracts important features from images such as edges,textures,shapes and patterns using filters
2.ReLU:reduces non-linearity into network,speeds up training,reduces vanishing gradient problem        
3.Batch Normalization:values after convolution ,Stablizies training
,improves model performance
4.Maxpooling reduces the spatial size of feature maps
5.Flatten Layer convers 2D feature maps into a 1D  vector
6.fully connected layers combine all extracted features and perform final classification &.Droupout randomly disables some neurons during training 8.The final layer produces outputs equal to the number of classes
